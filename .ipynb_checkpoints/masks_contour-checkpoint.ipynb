```python
brains = []
```

```python
folders = os.listdir(path)
for folder in folders:
    files = os.listdir(os.path.join(path, folder,'png'))
    for file in files:
        if file[-6:-4] == '12':
            brains.append(cv2.imread(os.path.join(path,folder,'png',file),3))
```

```python
im2,contours,hierarchy = cv2.findContours(masks[9], cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

for contour in contours:
     cv2.drawContours(brains[9], contour, -1, (255,0,0), thickness = 1)

b,g,r = cv2.split(brains[9])

r = cv2.add(b, 30, dst = b, mask = masks[9], dtype = cv2.CV_8U)

cv2.merge((b,g,r), brains[9]);
```

```python
plt.figure(figsize = (10,10))
plt.imshow(brains[9])
```
